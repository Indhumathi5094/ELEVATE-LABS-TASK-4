```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
data=pd.read_csv('/content/data.csv')
```

```python
x=pd.DataFrame(data.drop(['diagnosis', 'Unnamed: 32'],axis=1))
y=pd.Series(data.diagnosis)
```

```python
x.shape
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_prob=model.predict_proba(x_test)[:,1]
```

```python
from sklearn.metrics import confusion_matrix,classification_report, roc_auc_score, roc_curve
cm=confusion_matrix(y_pred,y_test)
cm
```

```python
cr=classification_report(y_pred,y_test)
cr
```

```python
y_test_numeric = y_test.map({'B': 0, 'M': 1})
y_pred_numeric = pd.Series(y_pred).map({'B': 0, 'M': 1})

ras=roc_auc_score(y_test_numeric,y_pred_numeric)
ras
```

```python
fpr, tpr, thresholds = roc_curve(y_test_numeric, y_pred_numeric)
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {ras:.2f})")
plt.plot([0,1], [0,1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
threshold = 0.6
y_pred_thresh = (y_prob >= threshold).astype(int)
th=confusion_matrix(y_test_numeric, y_pred_thresh)
th
```
